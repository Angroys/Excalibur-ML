```python
from sklearn.model_selection import train_test_split
x = df[["Tumor_Size"]]

y = df['Malignant']

x_train, x_val, y_train, y_val = train_test_split(x, y, test_size=0.5, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train, y_train)
model.predict(x_val)
```

```python
y_pred = model.predict_proba(x_val)[:, 1]
```

```python
y_pred
```

```python
import numpy as np
tumor_size_range = np.linspace(0,5,200)
```

```python
tumor_size_range = tumor_size_range.reshape(-1, 1)
predicted_probabilities = model.predict_proba(tumor_size_range)[:, 1]
```

```python
import matplotlib.pyplot as plt

plt.plot(tumor_size_range, predicted_probabilities)

plt.scatter(x_val[y_val== 0], y_val[y_val==0], color='blue', label='0', marker='x')
plt.scatter(x_val[y_val== 1], y_val[y_val==1], color='red', label='1', marker='x')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import accuracy_score
thresholds=np.linspace(0,1,20)
scores = []
for t in thresholds:
    score = accuracy_score(y_val, y_pred >=t)
    scores.append(score)

plt.plot(thresholds, scores)
plt.xlabel("Thresholds")
plt.ylabel("Accuracy")
```

```python
actual_pos = (y_val == 1)
actual_neg = (y_val == 0)
```

```python
pred_pos = (y_pred >=0.5)
pred_neg = (y_pred< 0.5)
```

```python
tp = (actual_pos & pred_pos).sum()
tn = (actual_neg & pred_neg).sum()
fp = (actual_neg & pred_pos).sum()
fn = (actual_pos & pred_neg).sum()
```

```python
#Rata de clasificare a clasei pozitive pe care noi le-am putut clasifica
recall = tp / (tp+fn)

precision = tp/(tp+fp)
```

```python
precision
```

```python
from sklearn.metrics import f1_score

def f1_score_custom(recall, precision):
    return 2*precision*recall/(precision + recall) 
```

```python
f1 = f1_score_custom(recall, precision)
f1
```

```python
from sklearn.metrics import f1_score, recall_score, precision_score
thresholds=np.linspace(0,1,20)
scores = []
for t in thresholds:
    recall = recall_score(y_val, y_pred>=t)
    precision = precision_score(y_val, y_pred>=t)
    scores.append((t, recall, precision))

scores_f = pd.DataFrame(scores, columns=["thresholds", 'recall', 'precision'])
plt.plot(scores_f.thresholds, scores_f.precision, label = "precision")
plt.plot(scores_f.thresholds, scores_f.recall, label = "precision")

plt.legend()
```

```python
from sklearn.metrics import f1_score, recall_score, precision_score
thresholds=np.linspace(0,1,20)
scores = []
for t in thresholds:
    pred_pos = (y_pred >=t)
    pred_neg = (y_pred< t)
    tp = (actual_pos & pred_pos).sum()
    tn = (actual_neg & pred_neg).sum()
    fp = (actual_neg & pred_pos).sum()
    scores.append((t, tp,tn,fp, fn ))

scores_f = pd.DataFrame(scores, columns=["thresholds", 'tp', 'tn', 'fp', 'fn'])

scores_f['tpr'] = scores_f["tp"]/(scores_f["tp"]+ scores_f["fn"])
scores_f['fpr'] = scores_f["fp"]/(scores_f["fp"]+ scores_f["tn"])
```
