```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = [1,2,3,4,5]
y = [10,11,12,13,14]

plt.plot(x, y)
plt.xlabel("ACEASTA ESTE SCRIS PENTRU X")
plt.ylabel("ACEASTA ESTE SCRIS PENTRU Y")
plt.title("Aceasta este un titlu")
plt.show()
```

```python
categorii = ["A", "B", "C"]
ceva = [1,2,3]
plt.bar(categorii, ceva)
```

```python
data = [1,2,2,3,3,3,4,4,4,4]

plt.hist(data, bins = 4, color="cyan")
plt.show()
```

```python
size = [10,20,20,30,50,5]
labels = ["a", "b","c", "d", "e", "f"]

plt.pie(size, labels=labels, autopct="%1.1f%%")
plt.show()
```

```python
x = np.linspace(0,10, 100)
y1= x
y2= x **2
fig,axs = plt.subplots(1,3, figsize=([12,4]))

axs[0].plot(x, y1)
axs[1].plot(x, y2)
axs[2].plot(x, np.sqrt(x))


plt.tight_layout()

plt.show()
```

```python
fig = plt.figure(figsize=(10, 8))

ax1 = plt.subplot2grid((2,2), (0,0))
ax2 = plt.subplot2grid((2,2), (0, 1))
ax3 = plt.subplot2grid((2,2), (1, 0), colspan=2)

ax1[0].plot(x, y1)
ax2[1].plot(x, y2)
ax3[2].plot(x, np.sqrt(x))

plt.show()
```
